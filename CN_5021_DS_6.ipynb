```python
# Importing required libraries

import numpy as np
import pandas as pd
```

**QN.1**

```python
# QN 1: Create a simple DataFrame with columns city, year, and pop

data = pd.DataFrame({
    'city': ['Bournemouth', 'London', 'Bristol', 'Manchester', 'Liverpool', 'Bournemouth', 'Leeds'],
    'year': [2020, 2021, 2020, 2021, 2022, 2022, 2023],
    'pop': [195000, 9000000, 463400, 553230, 498042, 196000, 789194]
})

# Display the DataFrame
data
```

**QN.2**

```python
# QN 2: Display the first 5 and last 5 rows of the DataFrame

print("First 5 rows of data:")
display(data.head())

print("\nLast 5 rows of data:")
display(data.tail())
```

```python
# QN 3: Filter rows where city is Bournemouth

bournemouth_data = data[data['city'] == 'Bournemouth']
bournemouth_data
```

**QN.7**

```python
# QN 7: Create a DataFrame of numbers between 0 and 200 (step of 5)
# We will reshape this array into a matrix for DataFrame creation

numbers = np.arange(0, 200, 5)
numbers = numbers.reshape(8, 5)   # Reshape to 8 rows and 5 columns

numbers_df = pd.DataFrame(numbers, columns=['C1', 'C2', 'C3', 'C4', 'C5'])
numbers_df
```

**QN.8**

```python
# QN 8: Demonstrate .loc and .iloc usage

# Using .iloc to select 2nd row, 3rd column (index 1,2)
iloc_example = numbers_df.iloc[1, 2]

# Using .loc to select using label-based indexing
loc_example = numbers_df.loc[3, 'C4']

print("Using .iloc -> Row 2, Column 3:", iloc_example)
print("Using .loc  -> Row 4, Column C4:", loc_example)
```

**QN.9**

```python
# QN 9: Change the content of the DataFrame at row index 1 and column index 1

print("Before change:")
display(numbers_df)

# Change element at position [1,1]
numbers_df.iloc[1, 1] = 999

print("After change:")
display(numbers_df)
```

**QN.10**

```python
# QN 10: Use Boolean indexing to show values > 100

greater_than_100 = numbers_df[numbers_df > 100]
greater_than_100
```

**QN 11**

```python
# QN 11: Create a Boolean mask for numbers greater than 100

mask = numbers_df > 100
mask
```

**QN 12**

```python
# QN 12: Demonstrate slicing on numbers_df

# Slice first 3 rows and first 3 columns
sliced_df = numbers_df.iloc[:3, :3]
sliced_df
```
