**Question 1: Import libraries, load dataset, and display the dataset (3 marks)**

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Define the file path
file_path = 'airquality.csv'

# Load the dataset, selecting only the relevant columns (Date and the three gas readings)
# Based on the CSV structure: [0:Date, 1:NO, 2:Status, 3:NO2, 4:Status, 5:NOx, 6:Status]
# We select indices 0, 1, 3, and 5
df = pd.read_csv(file_path, usecols=[0, 1, 3, 5])

# Rename columns to match the question paper and set 'Date' as the index
df.columns = ['Date', 'Nitric oxide', 'Nitrogen dioxide', 'Nitrogen oxides as nitrogen dioxide']
df = df.set_index('Date')

# Display the resulting DataFrame
print("--- Initial Loaded and Cleaned Dataset ---")
df
```

**Question 2: How many rows of data does this dataset have? (1 mark)**

```python
# Calculate the number of rows (length of the DataFrame)
num_rows = len(df)

# Print the result
print(f"The dataset has {num_rows} rows of data.")
```

**Question 3: Write code to show the first 35 rows of this dataset (1 mark)**

```python
# Display the first 35 rows using the .head() method
print("--- First 35 Rows of the Dataset ---")
df.head(35)
```

**Question 4: Write code to show only the rows with missing values for Nitric oxide. (3 marks)**

```python
# Filter the DataFrame to show only rows where 'Nitric oxide' is missing (NaN)
missing_nitric_oxide = df[df['Nitric oxide'].isnull()]

# Display the resulting DataFrame
print("--- Rows with Missing Values for Nitric oxide ---")
missing_nitric_oxide
```

**Question 5: How many rows of this dataset do not have any data for Nitric oxide? (1 mark)**

```python
# Count the number of missing values (NaNs) in the 'Nitric oxide' column
count_missing_no = df['Nitric oxide'].isnull().sum()

# Print the result
print(f"Number of rows with missing data for Nitric oxide: {count_missing_no}")
```

**Question 6: Write code to show only the rows with missing values for Nitrogen oxides as nitrogen dioxide. (3 marks)**

```python
# Filter the DataFrame to show only rows where 'Nitrogen oxides as nitrogen dioxide' is missing (NaN)
missing_nox = df[df['Nitrogen oxides as nitrogen dioxide'].isnull()]

# Display the resulting DataFrame
print("--- Rows with Missing Values for Nitrogen oxides as nitrogen dioxide ---")
missing_nox
```

**Question 7: How many rows of this dataset do not have any data for Nitrogen oxides as nitrogen dioxide? (1 mark)**

```python
# Count the number of missing values (NaNs) in the 'Nitrogen oxides as nitrogen dioxide' column
count_missing_nox = df['Nitrogen oxides as nitrogen dioxide'].isnull().sum()

# Print the result
print(f"Number of rows with missing data for Nitrogen oxides as nitrogen dioxide: {count_missing_nox}")
```

**Question 8: Write code to fill in the missing Nitric oxide values with the average value for Nitric oxide. You need to fill in only the column for Nitric oxide. Also include code for displaying the dataset. (4 marks)**

```python
# 1. Calculate the average (mean) of the existing 'Nitric oxide' values
mean_nitric_oxide = df['Nitric oxide'].mean()

# Print the calculated mean value
print(f"Calculated average value for Nitric oxide: {mean_nitric_oxide:.4f}")

# 2. Fill the missing values in the 'Nitric oxide' column with the calculated mean
# The inplace=True parameter updates the DataFrame directly
df['Nitric oxide'].fillna(mean_nitric_oxide, inplace=True)

# 3. Display the updated dataset
print("\n--- Dataset after filling missing Nitric oxide values with the mean ---")
df
```

**Question 9: Write code to display the first 35 rows of the dataset again (1 mark)**

```python
# Display the first 35 rows of the updated DataFrame
print("--- First 35 Rows of the UPDATED Dataset ---")
df.head(35)
```

**Question 10: Write code to drop all rows and columns missing at least one value. Ensure that you update the dataframe and display it (2 marks)**

```python
# 1. Drop all rows that still have any missing value (NaN)
# This will remove rows where Nitrogen dioxide or Nitrogen oxides were missing.
df_cleaned = df.dropna(how='any', axis=0)

# 2. Drop all columns that still have any missing value (NaN) - redundant after step 1, but included for completeness.
df_cleaned = df_cleaned.dropna(how='any', axis=1)

# 3. Update the original DataFrame reference (as required by the question)
df = df_cleaned

# 4. Display the resulting DataFrame and its final shape
print("--- Dataset after dropping all rows and columns with missing values ---")
print(f"Final shape of the DataFrame: {df.shape}")
df
```
